## ðŸ“˜ What is Machine Learning?

Machine learning (ML) is a subfield of artificial intelligence (AI) that gives computers the ability to learn from data and make predictions or decisions without being explicitly programmed.

### Types of Machine Learning

1. **Supervised Learning** â€“ Trained with labeled data  
2. **Unsupervised Learning** â€“ Trained with unlabeled data  
3. **Semi-supervised Learning** â€“ Mix of both labeled and unlabeled data  
4. **Reinforcement Learning** â€“ Learns by trial and error via reward signals

```python

from sklearn.datasets import load_iris, fetch_california_housing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
iris = load_iris(as_frame=True)
iris_df = iris.frame

# Preview the Iris dataset
iris_df.head()
```

```python

sns.pairplot(iris_df, hue='target', palette='husl')
plt.suptitle("Iris Dataset - Pairwise Features", y=1.02)
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

X = iris_df.drop(columns='target')
y = iris_df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python

from sklearn.linear_model import LinearRegression

housing = fetch_california_housing(as_frame=True)
housing_df = housing.frame

X = housing_df[['MedInc']]
y = housing_df['MedHouseVal']

model = LinearRegression()
model.fit(X, y)

plt.scatter(X, y, alpha=0.3)
plt.plot(X, model.predict(X), color='red')
plt.title('California Housing: Income vs House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```
